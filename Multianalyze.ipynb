```python
def multi_analyze(df, tracks, n):
    #chroms = tracks[0].chrom.unique()
    Es = tracks[0].columns.tolist()[5::n]
    results = dict()
    for E in Es:
        compared_df = tracks[0].iloc[:, :5]
        for i in range(len(df)):
            cell_line = df.iloc[i].cell_line
            compared_df[cell_line] = tracks[i][E]
        results[E] =compared_df
    return results
            
        
        
```

```python
results = multi_analyze(df, tracks, n)
```

```python
def get_correlations(resdict, title):
    
    plt.figure(1, figsize  =(15, 15))
    plt.suptitle(title)
    i = 1
    for k, v in resdict.items():
        plt.subplot(3,3,i)
        corr = v.iloc[:,5:].corr()
        #print(corr)
        plt.imshow(corr,cmap='coolwarm', vmin=0, vmax=1)
        plt.colorbar();
        ticks = np.arange(0,len(corr.columns),1)
        plt.xticks(ticks, labels =corr.columns,  rotation=90)
        plt.yticks(ticks, labels = corr.columns)
        for z in range(len(corr)):
            for j in range(len(corr)):
                text = plt.text(j, z, np.round(corr.iloc[z, j], 2),
                       ha="center", va="center", color="w")
        plt.title(k)
        i+=1
    plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.5,
                    wspace=0.35)
```

```python
get_correlations(results, 'Cell line E1 correlations')
```

```python
def get_chrom_correlations(resdict, chroms):
    
    for ch in chroms:
        plt.figure(figsize  =(30, 5))
        plt.suptitle(ch)
        i = 1
        for k, v in resdict.items():
            plt.subplot(1,5,i)
            region = v[v.chrom == ch]
            corr = region.iloc[:,5:].corr()
            #print(corr)
            plt.imshow(corr,cmap='coolwarm', vmin=0, vmax=1)
            plt.colorbar();
            ticks = np.arange(0,len(corr.columns),1)
            plt.xticks(ticks, labels =corr.columns,  rotation=90)
            plt.yticks(ticks, labels = corr.columns)
            plt.title(k)
            i+=1
        plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.5,
                    wspace=0.35)
```

```python
chroms = tracks[0].chrom.unique()
```

```python
get_chrom_correlations(results, chroms)
```
